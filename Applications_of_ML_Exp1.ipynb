```python
import pandas as pd

# Load the Iris dataset
df = pd.read_csv("/content/Iris.csv")

df.loc[df['Id'] % 5 == 0, 'SepalLengthCm'] = None


missing_values = df.isnull().sum()
print("Missing Values:")
print(missing_values)


df['SepalLengthCm'].fillna(df['SepalLengthCm'].mean(), inplace=True)

df['PetalRatio'] = df['PetalLengthCm'] / df['PetalWidthCm']

df_encoded = pd.get_dummies(df, columns=['Species'], prefix='Species')

df_encoded.columns = df_encoded.columns.str.strip()


```

```python
import pandas as pd


data1 = {'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']}
data2 = {'ID': [2, 3, 4], 'Age': [25, 30, 22]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)


print("Dataset 1:")
print(df1)

print("\nDataset 2:")
print(df2)


inner_join = pd.merge(df1, df2, on='ID', how='inner')
print("\nInner Join:")
print(inner_join)


outer_join = pd.merge(df1, df2, on='ID', how='outer')
print("\nOuter Join:")
print(outer_join)


left_join = pd.merge(df1, df2, on='ID', how='left')
print("\nLeft Join:")
print(left_join)


right_join = pd.merge(df1, df2, on='ID', how='right')
print("\nRight Join:")
print(right_join)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/Iris.csv")


df['Species'].value_counts().plot(kind='bar', color=['skyblue', 'lightgreen', 'lightcoral'], edgecolor='black')
plt.title('Bar Plot: Species Counts')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()

df.plot(x='Id', y='SepalLengthCm', kind='line', marker='o', color='blue')
plt.title('Line Plot: Sepal Lengths')
plt.xlabel('Id')
plt.ylabel('Sepal Length (cm)')
plt.show()

df.plot(x='SepalLengthCm', y='SepalWidthCm', kind='scatter', color='green', marker='o')
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()


correlation_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()

highly_correlated = correlation_matrix.abs() > 0.7
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5, mask=~highly_correlated)
plt.title('Highly Correlated Features (Absolute Correlation > 0.7)')
plt.show()


df.hist(figsize=(10, 8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Histograms for Numerical Columns', y=1.02)
plt.show()

df.plot(kind='box', figsize=(10, 8), color='lightgreen', vert=False)
plt.title('Box Plots for Numerical Columns')
plt.show()
```

```python
import numpy as np

arr = np.arange(1, 11)

arr2 = np.arange(11, 21)

add_result = arr + arr2
sub_result = arr - arr2
mul_result = arr * arr2
div_result = arr / arr2

print("arr + arr2 =", add_result)
print("arr - arr2 =", sub_result)
print("arr * arr2 =", mul_result)
print("arr / arr2 =", div_result)
```

```python
import numpy as np


arr = np.arange(1, 11)
arr_2x5 = arr.reshape(2, 5)


transposed_matrix = arr_2x5.T

flattened_array = transposed_matrix.flatten()


arr2 = np.arange(11, 21)
stacked_result = np.vstack((arr, arr2))


print("1. Reshaped 'arr' into a 2x5 matrix:")
print(arr_2x5)

print("\n2. Transposed matrix:")
print(transposed_matrix)

print("\n3. Flattened the transposed matrix into a 1D array:")
print(flattened_array)

print("\n4. Stacked 'arr' and 'arr2' vertically:")
print(stacked_result)
```
